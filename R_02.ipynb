# Load df.unnest

```r
print("Loading df.unnest ...")
df.unnest <- readRDS("/ahg/regevdata/projects/ICA_Lung/Akhil/projects/prev/mutation_tracking/metadata.manipulated.unnested.rds")
print("Loaded df.unnest successfully!")
dim(df.unnest)
head(df.unnest)
```

```r
print("Started subsetting...")
df.unnest.fil <- df.unnest %>%
    select("Collection date", "AA_Substitutions", "Accession ID", "Location") %>%
    rename("collection_date" = "Collection date", "aa_subs" = "AA_Substitutions", "accession_id" = "Accession ID", "location" = "Location") %>%
    mutate(across(location, as.factor))
print("Started changing the format...")
df.unnest.fil$collection_date <- as_date(df.unnest.fil$collection_date)

dim(df.unnest.fil)
head(df.unnest.fil)
```

# DF1 and DF2

```r
# Assuming you have loaded the required libraries and generated the output as shown
print("Started creating DF1...")
DF1 <- df.unnest.fil %>%
    mutate(region = str_split(string = aa_subs, pattern = "_")) %>%
    mutate(mutation_name = sapply(region, function(x) x[2]), region = sapply(region, function(x) x[1]))

dim(DF1)
head(DF1)

print("Started saving DF1...")
saveRDS(DF1, "/ahg/regevdata/projects/ICA_Lung/Akhil/projects/prev/mutation_tracking/r_analysis/rds_files/DF1.rds")
print("Saved DF1 successfully!")
```

```r
print("Started creating DF2...")
DF2 <- df.unnest.fil %>%
    group_by(aa_subs) %>%
    summarise(counts = n()) %>%
    ungroup() %>%
    arrange(desc(counts))

dim(DF2)
head(DF2)
print("Started saving DF2...")
saveRDS(DF2, "/ahg/regevdata/projects/ICA_Lung/Akhil/projects/prev/mutation_tracking/r_analysis/rds_files/DF2.rds")
print("Saved DF2 successfully!")
```

## Mutating DF1

```r
DF1
```

```r
nseqs <- length(unique(DF1$accession_id))
nseqs
```

```r
print("Started computing DF1.unique ...")
DF1.unique <- DF1 %>%
  arrange(collection_date) %>%
  distinct(aa_subs, .keep_all = TRUE)

dim(DF1.unique)
head(DF1.unique)

print("Started saving DF1.unique ...")
saveRDS(DF1.unique, "/ahg/regevdata/projects/ICA_Lung/Akhil/projects/prev/mutation_tracking/r_analysis/rds_files/DF1.unique.rds")
print("Saved DF1.unique successfully!")
```

```r
sort(table(DF1.unique$region), decreasing = T)
```

```r
# Removing the mutation which has no info on aa_subs
DF1.unique <- DF1.unique[-which(DF1.unique$region == ""),]
DF1.unique
```

```r
DF1.unique <- DF1.unique %>%
    mutate(
    ref = sub("^(.*[A-Za-z])(\\d+)([A-Za-z].*)$", "\\1", mutation_name),
    pos = as.integer(sub("^(.*[A-Za-z])(\\d+)([A-Za-z].*)$", "\\2", mutation_name)),
    alt = sub("^(.*[A-Za-z])(\\d+)([A-Za-z].*)$", "\\3", mutation_name)
  )
DF1.unique
```

```r
DF1.unique <- DF1.unique[!(is.na(DF1.unique$collection_date)), ]
DF1.unique
```

```r
DF1.unique
```

```r
print("Started saving DF1.unique.mutated ...")
saveRDS(DF1.unique, "/ahg/regevdata/projects/ICA_Lung/Akhil/projects/prev/mutation_tracking/r_analysis/rds_files/DF1.unique.mutated.rds")
print("Saved DF1.unique.mutated successfully!")
```

## Mutating DF2

```r
DF2 <- DF2 %>%
    mutate(region = str_split(string = aa_subs, pattern = "_")) %>%
    mutate(mutation_name = sapply(region, function(x) x[2]), region = sapply(region, function(x) x[1])) %>%
    mutate(fraction = counts / nseqs) %>%
    mutate(per_abundance = fraction * 100)

DF2
```

```r
# remove 1392 mutations with missing collection_date
DF2 <- filter(DF2, aa_subs %in% DF1.unique$aa_subs)
DF2
```

```r
sort(table(DF1.unique$region), decreasing = T)
sort(table(DF2$region), decreasing = T)
sum(table(DF2$region))
identical(table(DF1.unique$region), table(DF2$region))
```

```r
# Very low per_abundance of 1392 mutations with missing collection dates
filter(DF2, aa_subs %in% DF1.unique[is.na(DF1.unique$collection_date),]$aa_subs) %>%
    arrange(desc(per_abundance))
```

```r
DF2[which(DF2$region != ""),]
```

```r
# Removing the mutation which has no info on aa_subs
DF2 <- DF2[which(DF2$region != ""), ]
DF2
```

```r
DF2 <- DF2 %>%
    mutate(
    ref = sub("^(.*[A-Za-z])(\\d+)([A-Za-z].*)$", "\\1", mutation_name),
    pos = as.integer(sub("^(.*[A-Za-z])(\\d+)([A-Za-z].*)$", "\\2", mutation_name)),
    alt = sub("^(.*[A-Za-z])(\\d+)([A-Za-z].*)$", "\\3", mutation_name)
  )
DF2
```

```r
DF2.subs <- DF2 %>% filter(ref != "ins") %>%
    filter(ref != "del") %>%
    filter(alt != "ins") %>%
    filter(alt != "del")
DF2.subs
```

```r
DF2.del <- filter(DF2, ref == "del" | alt == "del")
DF2.del
```

```r
DF2.ins <- filter(DF2, ref == "ins" | alt == "ins")
DF2.ins
```

```r
nrow(DF2) == nrow(DF2.subs) + nrow(DF2.del) + nrow(DF2.ins)
```

```r
print("Started saving DF2...")
saveRDS(DF2, "/ahg/regevdata/projects/ICA_Lung/Akhil/projects/prev/mutation_tracking/r_analysis/rds_files/DF2.mutated.rds")
print("Saved DF2 successfully!")
```
